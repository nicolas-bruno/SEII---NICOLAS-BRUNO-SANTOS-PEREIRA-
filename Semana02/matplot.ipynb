```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
x = [1,2,3]
y = [2,4,6]
plt.plot (x,y)
plt.title = ('grafico exemplo')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
x = [1,2,3]
y = [2,4,6]
plt.plot (x,y,label ='2x',color='green',linewidth = 6, markersize = 10)
plt.title = ('grafico exemplo')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
x = [1,2,3]
y = [2,4,6]
x2 = np.arange(0,4.5,0.5)
plt.plot (x2,x2**2,label ='2x',color='green',linewidth = 6, markersize = 10)
plt.title = ('grafico exemplo')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.show()
```
